```python
import utils
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')

# Load data
df: pd.DataFrame = utils.load_original_data()

df.head()
```

#### Summary Statistics

```python
df.describe().T
```

This dataset contains 2126 records of 21 features extracted from Cardiotocogram exams, which were then classified into 3 classes           
(represented in fetal_health target variable):
- 1 - Normal
- 2 - Suspect
- 3 - Pathological

#### Features Visualization and Analysis

```python
# boxplot for outliers detection
fig = plt.figure(figsize=(15, 15))
for i, col in enumerate(df.columns):
    fig.add_subplot(4, 6, i+1)
    df.boxplot(column=col)
plt.show()
```

```python
# Features distribution 
fig = plt.figure(figsize=(15, 15))
df.hist(ax=fig.gca())
plt.show()
```

```python
# Target Evaluation to find out if the data is imbalanced or not

colours=["green", "blue", "red"]
sns.countplot(data= df, x="fetal_health",palette=colours)
```

- The count plot of targets indicates an imbalance in data. This is a case that tends to provide misleading classification accuracy.

```python
# correlation matrix with seaborn

corr = df.corr()
fig = plt.figure(figsize=(15, 15))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

- Based on the correlation matix, it is possible to observe that "accelerations","prolongued_decelerations", "abnormal_short_term_variability" and "percentage_of_time_with_abnormal_long_term_variability" are the features with higher correlation with fetal_health. Let's further analyze these with percentage_of_time_with_abnormal_long_term_variability on y axis in order to provide insights into fetus health. 

#### Bivariate Analysis

```python
# Accelerations Vs Fetal Movement by Fetal Health

sns.lmplot(data =df,x="accelerations",y="percentage_of_time_with_abnormal_long_term_variability",palette=colours, hue="fetal_health",legend_out=False)
plt.show()
```

```python
# Prolongued Decelerations Vs Fetal Movement by Fetal Health

sns.lmplot(data =df,x="prolongued_decelerations",y="percentage_of_time_with_abnormal_long_term_variability",palette=colours, hue="fetal_health",legend_out=False)
plt.show()
```

```python
# Abnormal Short Term Variability Vs Fetal Movement by Fetal Health

sns.lmplot(data =df,x="abnormal_short_term_variability",y="percentage_of_time_with_abnormal_long_term_variability",palette=colours, hue="fetal_health",legend_out=False)
plt.show()
```
